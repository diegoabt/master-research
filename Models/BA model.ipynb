### Barabási-Albert Model
We begin with a simple test. First we generate the network under the known model and then we plot the results. 
The command receives three inputs: the number of nodes of the graph, _________ and _________. Since our research focus on trees, it is important to keep the second input equal to 1.

```python
ba=nx.barabasi_albert_graph(12,1,0) #IMPORTANT PART! second input must be 1
nx.draw_networkx(ba, edge_color='k', with_labels=True, alpha=.4, linewidths=0)
plt.axis('off')
plt.show()
```
