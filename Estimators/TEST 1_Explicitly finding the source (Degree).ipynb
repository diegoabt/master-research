```python
import time
import networkx as nx
import powerlaw
import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np
from plotly import tools
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot,iplot
print (__version__) # requires version >= 1.9.0s

#Always run this the command before at the start of notebook
init_notebook_mode(connected=True)
import plotly.graph_objs as go
```

## Degree
### BA

```python
resultsDG={} #This dict contains all the vrtices found by the rl estimator for different graph sizes
values_dg={} #This dict is created from resultsDC
results_dg_time={}
```

```python
#
for N in [100,200,300,400,500]:
    results_dg=[]
    start_time=time.time()
    for i in range(0,500):
        ba=nx.barabasi_albert_graph(N,1)
        dc=nx.degree_centrality(ba)
        list_maximum=[]
        center=max(dc.iteritems(), key=operator.itemgetter(1))[1]
        for i in range(N):
                if dc[i]==center:
                    list_maximum.append([i,dc[i]])
        pair=random.choice(list_maximum)
        results_dg.append([N,pair[0]])
    final_time= -start_time +time.time()
    results_dg_time[str(N)]=final_time
    
    resultsDG[str(N)]=[i[1] for i in results_dg]
    print N
    

data_1={}
data_2={}
fig = tools.make_subplots(rows=5, cols=2, subplot_titles=('100', '100',
                                                          '200', '200',
                                                         '300','300',
                                                         '400','400',
                                                         '500','500'))
j=0
for i in [100,200,300,400,500]:
    j+=1
    
    #SIMPLEST HISTOGRAM
    x = resultsDG[str(i)]
    data_1[str(i)] = go.Histogram(x=x,
                                  marker=dict(
        color='dodgerblue'
    ),
                     xbins=dict(
            start=0,
            end=i,
            size=1.
        ))
    fig.append_trace(data_1[str(i)], j, 1)
    #CUMM HISTOGRAM
    data_2[str(i)] = go.Histogram(x=x,
                                  marker=dict(
        color='royalblue'
    ),
                     xbins=dict(
        start=0,
        end=i,
        size=1.
    ),
                     histnorm='probability',
                     cumulative=dict(enabled=True))
    fig.append_trace(data_2[str(i)], j, 2)
    
```

```python
fig['layout'].update(height=1000, width=800, title='Single vertex: Degree (BA)')
iplot(fig, filename='simple-subplot')
```
