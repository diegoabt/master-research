### Degree center 

```python
resultsDC={} #This dict contains all the vrtices found by the dc estimator for different graph sizes
values_dc={} #This dict is created from resultsDC
results_dc_time={}
```

```python
graph_size=[100,200,300,400,500,600,700,800,900,1000,1500,2000,2500,3000,4000,5000,10000]


for N in graph_size:
    results_dc=[]
    start_time=time.time()
    for i in range(0,500):
        ba=nx.barabasi_albert_graph(N,1)
        dc=nx.degree_centrality(ba)
        center=max(dc.iteritems(), key=operator.itemgetter(1))[0]
        results_dc.append([N,center])
    final_time= -start_time +time.time()
    results_dc_time[str(N)]=final_time
    
    resultsDC[str(N)]=[i[1] for i in results_dc ]
    %matplotlib inline
    values, bins, _ =plt.hist(resultsDC[str(N)], bins=range(0,N),color='m',normed=True)
    values_dc[str(N)]=list(values)
    plt.ylabel('');
    plt.xlabel('dc center')
    print N
```

Exporting the results

```python
degcen_results = open('degcenSim500.txt', 'w')
for i in [100,200,300,400,500,600,700,800,900,1000,1500,2000,2500,3000,4000,5000,10000]:
  print>>degcen_results, str(i),resultsDC[str(i)]
degcen_results.close()
```

##### Confidence sets function

```python
def confset_dc(l):
    sred = sorted(l.items(), reverse=True, key=lambda value: value[1])
    #sorted dc values. first element has the maximum pr value
    index=0
    for i in range(0,len(sred)):
        if sred[i][0]==0:
            return index
            break
        index=index+1
```

##### Confidence sets function

```python
def confset_cc(l):
    sred = sorted(l.items(), reverse=True, key=lambda value: value[1])
    #sorted dc values. first element has the maximum pr value
    index=0
    for i in range(0,len(sred)):
        if sred[i][0]==0:
            return index
            break
        index=index+1
```
